```python
#Create A Plot with 9 subplots-line
# -stack
# -Box plot
# -histogram
# -bar chart
# -violin chart
# -Horizontal bar chart
# -sactter plot
# -Area plot
# ---
# ---
# ---
#     with x axis, y label and title
#     -label on bar
#     -gab between the plot( gap kaise increase , decrease krte hai pta hona chahiye)


import matplotlib.pyplot as plt
import numpy as np
a= np.linspace(0,10,5)
b=[1,5,1,4,2,3,6,5,4,7,8,9,5,4,1,2,3,8,7,4,5]
x= [2,4,6,8,10]
y= [10,15,20,40,25]
z= [[1,4,3,2,1],
    [5,4,3,2,3],
    [3,2,6,9,1]] 

ax1= plt.subplot(331)
ax2=plt.subplot(332)
ax3=plt.subplot(333)
ax4=plt.subplot(334)
ax5= plt.subplot(335)
ax6=plt.subplot(336)
ax7= plt.subplot(337)
ax8=plt.subplot(338)
ax9= plt.subplot(339)


ax1.plot(a , marker= 'o' )
ax1.set_title('Line Plot')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')

ax2.stackplot(x,z)
ax2.set_title('Stack Plot')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')

ax3.boxplot(x)
ax3.set_title('Box Plot')
ax3.set_xlabel('X')
ax3.set_ylabel('Y')

ax4.hist(b ,bins = 5, color= 'b', edgecolor= 'yellow')
ax4.set_title('Histogram Plot')
ax4.set_xlabel('X')
ax4.set_ylabel('Y')

ax5.bar(x,y, color =['blue','yellow'])
ax5.set_title('Bar Chart')
ax5.set_xlabel('X')
ax5.set_ylabel('Y')

ax6.violinplot(x)
ax6.set_title('Violin Plot')
ax6.set_xlabel('X')
ax6.set_ylabel('Y')

ax7.barh(x,y, color =['blue','yellow'])
ax7.set_title('Horizontal Bar Plot')
ax7.set_xlabel('X')
ax7.set_ylabel('Y')

ax8.scatter(x,y , color = 'r' , edgecolor = 'yellow',marker='^')
ax8.set_title('Scatter Plot')
ax8.set_xlabel('X')
ax8.set_ylabel('Y')

ax9.stackplot(x,y , color= 'b', edgecolor = 'yellow')
ax9.set_title('Area Plot')
ax9.set_xlabel('X')
ax9.set_ylabel('Y')

plt.subplots_adjust(hspace =1.5, wspace=1)

plt.show()
```
