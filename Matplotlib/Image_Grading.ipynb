# Matplotlib

```python
#Project On Numpy---
import numpy as np
from PIL import Image   #PIL - python image library
import matplotlib.pyplot as plt
img = Image.open(r"D:\s3.jpg") #img = variable
img_arr = np.array(img)     #image changed into an array

plt.imshow(img_arr)
```

```python
plt.imshow(img_arr[0:200 , 0:350]) # for first quadrant
```

```python
plt.imshow(img_arr[0:200 , 350:]) # for second quadrant
```

```python
plt.imshow(img_arr[200: , 0:350 , 0]) # for third quadrant
```

```python
plt.imshow(img_arr[200: , 350:]) # for first quadrant
```

```python
#R - 0, G-1 B-2 ,,, To remove color
plt.imshow(img_arr[200: , 350: ,1])
```
